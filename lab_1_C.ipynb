```python
import random
import numpy as np

# Interpolation Search for Floating-Point Numbers
def interpolation_search(arr, key, epsilon=1e-9):
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= key <= arr[high]:
        comparisons += 1

        # Avoid division by zero
        if abs(arr[high] - arr[low]) < epsilon:
            if abs(arr[low] - key) < epsilon:
                return low, comparisons
            return -1, comparisons

        # Estimate position
        pos = low + int(
            ((key - arr[low]) * (high - low)) /
            (arr[high] - arr[low])
        )

        pos = max(low, min(pos, high))

        if abs(arr[pos] - key) < epsilon:
            return pos, comparisons

        elif arr[pos] < key:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparisons


# Experiment Function
def run_experiment(size, searches=100):
    # Generate uniformly distributed floating-point values
    arr = sorted(np.random.uniform(0.0, 1000.0, size))

    total_comparisons = 0

    for _ in range(searches):
        # Select a random existing element
        key = random.choice(arr)

        _, comps = interpolation_search(arr, key)
        total_comparisons += comps

    avg_comparisons = total_comparisons / searches
    return avg_comparisons


# Dataset sizes
sizes = [10000, 50000, 100000]

print("Interpolation Search Performance on Floating-Point Data")
print("-" * 60)

for size in sizes:
    avg = run_experiment(size)
    print(f"Dataset Size: {size:>6} | Average Comparisons: {avg:.2f}")
```
