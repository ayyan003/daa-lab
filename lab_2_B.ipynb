```python
def compute_lps(pattern):
    m = len(pattern)
    lps = [0] * m
    length = 0
    i = 1
    while i < m:
        if pattern[i] == pattern[length]:
            length += 1
            lps[i] = length
            i += 1
        else:
            if length != 0:
                length = lps[length - 1]
            else:
                lps[i] = 0
                i += 1
    return lps

pattern = "AABXAA"
text = "AABCAABXAAAZ"
lps = compute_lps(pattern)
print(f"Pattern: {pattern}")
print(f"LPS Array: {lps}")

def kmp_search(text, pattern, lps):
    n = len(text)
    m = len(pattern)
    i = 0 # index for text
    j = 0 # index for pattern

    steps = []
    while i < n:
        steps.append((i, j, text[i], pattern[j]))
        if pattern[j] == text[i]:
            i += 1
            j += 1

        if j == m:
            print(f"Pattern found at index {i - j}")
            j = lps[j - 1]
        elif i < n and pattern[j] != text[i]:
            if j != 0:
                j = lps[j - 1]
            else:
                i += 1
    return steps

steps = kmp_search(text, pattern, lps)
for step in steps:
    print(step)
```

```python
def kmp_trace(text, pattern):
    lps = compute_lps(pattern)
    n = len(text)
    m = len(pattern)
    i = 0
    j = 0

    print("Tracing text search:")
    while i < n:
        print(f"Comparing text[{i}]='{text[i]}' and pattern[{j}]='{pattern[j]}'")
        if text[i] == pattern[j]:
            i += 1
            j += 1
            if j == m:
                print(f"-> Match found at index {i - j}!")
                j = lps[j - 1]
        else:
            if j != 0:
                print(f"-> Mismatch! Shift pattern using lps[{j-1}] = {lps[j-1]}")
                j = lps[j - 1]
            else:
                print(f"-> Mismatch! Advance text index i to {i+1}")
                i += 1

kmp_trace(text, pattern)
```
